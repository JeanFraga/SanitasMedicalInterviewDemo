## Rank
Like sorting, but with collision detection

```python
dfp = df.sort_values('price', ascending=False)
dfp[['id', 'host_name', 'price']].head(3)
```

```python
# rank on price with min on dfp
dfp['price_rank'] = dfp.price.rank(ascending=False, method='min')

dfp[['id', 'host_name', 'price','price_rank']].head(3)
```
