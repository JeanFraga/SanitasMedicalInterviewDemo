```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Import CSV document from an online repository to explore Ratings of movies


```python
df = pd.read_csv('https://raw.githubusercontent.com/fivethirtyeight/data/master/inconvenient-sequel/ratings.csv')

df.tail(5)
```

This is an overview of the different categories present in this database

```python
df.dtypes
```

## This next line explores unique counts of the time stamp

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])
df['timestamp'].describe()
```

Since it is all unique we can set it as an index for the database

```python
df.set_index('timestamp', inplace=True)
df['2017-08-29']
```

## This next line segments to only keep IMDb users on the last day

```python
lastday = df['2017-08-29']
lastday_filtered = lastday[lastday['category'] == 'IMDb users']
lastday_filtered.tail(10)
```

```python
df['category'].value_counts()
```

## The next line plots a line graph that shows the number of respondents vs each date filtered by the last day

```python
lastday_filtered.respondents.plot()
```

```python
# A data point transposed so that it can be easily interpreted
final = lastday_filtered.tail(1)
final.T
```

```python
pct_columns = ['1_pct', '2_pct', '3_pct', '4_pct', '5_pct',
               '6_pct', '7_pct', '8_pct', '9_pct', '10_pct']
final[pct_columns].T
```

```python
plot_data = final[pct_columns].T
plot_data.index = range(1, 11)
plot_data
```

## This graph makes use of more styling to make the graph easier to look at.

# It measures the percent of total votes vs the rating they gave

```python
# Use the 538 stylesheet
plt.style.use('fivethirtyeight')

# generate our figure
fig = plt.figure()
fig.patch.set(facecolor='white')

# generate our axes (center section) for our plot
ax = plot_data.plot.bar(color="#ED713A", width=0.9)
ax.set(facecolor='white')
ax.patch.set_alpha(0)

# Labels the X and Y Axis
ax.set_ylabel("Percent of total votes", fontsize=9, fontweight='bold', 
              labelpad=10)
ax.set_xlabel("Rating", fontsize=9, fontweight='bold', labelpad=10)

# Asign a range to how sparse the data is plotted
ax.set_xticklabels(range(1,11), rotation=0)
ax.set_yticks(range(0,50,10))
ax.set_yticklabels(range(0, 50, 10))

# Remove Legend
ax.get_legend().remove()

# Displays the Graph
plt.show()
```
